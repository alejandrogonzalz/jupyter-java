```java
// Collections
import java.util.*;

List<String> languages = Arrays.asList("Java", "Python", "Kotlin", "Ruby");
languages.stream()
    .filter(lang -> lang.startsWith("J") || lang.startsWith("K"))
    .forEach(System.out::println);
```

```java
// Methods
int fibonacci(int n) {
    if (n <= 1) return n;
    return fibonacci(n - 1) + fibonacci(n - 2);
}

for (int i = 0; i < 10; i++) {
    System.out.printf("fib(%d) = %d%n", i, fibonacci(i));
}
```

```java
// Classes work too!
class Person {
    private String name;
    private int age;
    
    public Person(String name, int age) {
        this.name = name;
        this.age = age;
    }
    
    public String toString() {
        return name + " (" + age + " years old)";
    }
}

Person person = new Person("Alice", 30);
System.out.println(person);
```
